```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('train.csv')
```

```python
for col in train_data.drop(['id','num_sold'], axis=1).columns:
    frequency_table = train_data[col].value_counts()
    plt.show(sns.barplot(x=frequency_table.index, y=frequency_table.values))
```

```python
print('The number of dates analysed is ',len(train_data['date'].value_counts()))
print('Each date have 90 elements')
```
